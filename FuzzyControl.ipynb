```python
import numpy as np
import matplotlib.pyplot as plt     # for visualization

#Theta範圍
Txrange= np.arange(-(np.pi)/2,(np.pi)/2,0.0005) 
#Theta dot範圍
Tdxrange= np.arange(-np.pi,np.pi,0.001) 
#u範圍
uxrange= np.zeros(shape=201) 
uxrange[0] = -100
for i in range(1,201):
  uxrange[i] = uxrange[i-1] +1
```

```python
#membership function
def TNB(T):
  if T>=1.5:
    data=1
  else:
    data = (np.exp((-(T-1.5)**2)/0.1))
  return data

def TNS(T):
  data = (np.exp((-(T+0.5)**2)/0.1))
  return data

def TZR(T):
  data = (np.exp((-(T+0)**2)/0.1))
  return data

def TPS(T):
  data = (np.exp((-(T-0.5)**2)/0.1))
  return data

def TPB(T):
  if T<=-1.5:
    data=1
  else:
    data = (np.exp((-(T+1.5)**2)/0.1))
  return data

def TdNB(T):
  if T>=3:
    data=1
  else:
    data = (np.exp((-(T-3)**2)/1))
  return data

def TdNS(T):
  data = (np.exp((-(T+1.5)**2)/0.3))
  return data

def TdZR(T):
  data = (np.exp((-(T+0)**2)/0.3))
  return data

def TdPS(T):
  data = (np.exp((-(T-1.5)**2)/0.3))
  return data

def TdPB(T):
  if T<=-3:
    data=1
  else:
    data = (np.exp((-(T+3)**2)/1))
  return data


def uNB(T):
  if T>=90:
    data=1
  else:
    data = (np.exp((-(T-90)**2)/200))
  return data

def uNS(T):
  data = (np.exp((-(T-50)**2)/200))
  return data


def uZR(T):
  data = (np.exp((-(T-0)**2)/200))
  return data


def uPS(T):
  data = (np.exp((-(T+50)**2)/200))
  return data


def uPB(T):
  if T<=-90:
    data=1
  else:
    data = (np.exp((-(T+90)**2)/200))
  return data
```

```python
tnb = []
tns = []
tzr = []
tps = []
tpb = []

for i in np.arange(-(np.pi)/2,(np.pi)/2,0.0005):
  tnb.append(TNB(i))
  tns.append(TNS(i))
  tzr.append(TZR(i))
  tps.append(TPS(i))
  tpb.append(TPB(i))

tdnb = []
tdns = []
tdzr = []
tdps = []
tdpb = []

for i in np.arange(-np.pi,np.pi,0.001):
  tdnb.append(TdNB(i))
  tdns.append(TdNS(i))
  tdzr.append(TdZR(i))
  tdps.append(TdPS(i))
  tdpb.append(TdPB(i))

unb = []
uns = []
uzr = []
ups = []
upb = []

for i in range(-100,101):
  unb.append(uNB(i))
  uns.append(uNS(i))
  uzr.append(uZR(i))
  ups.append(uPS(i))
  upb.append(uPB(i))

plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis") 

plt.plot(Txrange,tnb,'g-')
plt.plot(Txrange,tns,'b-') 
plt.plot(Txrange,tzr,'c-') 
plt.plot(Txrange,tps,'r-') 
plt.plot(Txrange,tpb,'y-') 

plt.show()

plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis") 

plt.plot(Tdxrange,tdnb,'g-')
plt.plot(Tdxrange,tdns,'b-') 
plt.plot(Tdxrange,tdzr,'c-') 
plt.plot(Tdxrange,tdps,'r-') 
plt.plot(Tdxrange,tdpb,'y-') 

plt.show()

plt.xlabel("X-Axis") 
plt.ylabel("Y-Axis") 

plt.plot(uxrange,unb,'g-')
plt.plot(uxrange,uns,'b-') 
plt.plot(uxrange,uzr,'c-') 
plt.plot(uxrange,ups,'r-') 
plt.plot(uxrange,upb,'y-') 

plt.show()
```

```python
def control(T,Td):

  ternp = np.zeros(25)
  ternp[0] = TNB(T) * TdNB(Td)
  ternp[1] = TNB(T) * TdNS(Td)
  ternp[2] = TNB(T) * TdZR(Td)
  ternp[3] = TNB(T) * TdPS(Td)
  ternp[4] = TNB(T) * TdPB(Td)

  ternp[5] = TNS(T) * TdNB(Td)
  ternp[6] = TNS(T) * TdNS(Td)
  ternp[7] = TNS(T) * TdZR(Td)
  ternp[8] = TNS(T) * TdPS(Td)
  ternp[9] = TNS(T) * TdPB(Td)

  ternp[10] = TZR(T) * TdNB(Td)
  ternp[11] = TZR(T) * TdNS(Td)
  ternp[12] = TZR(T) * TdZR(Td)
  ternp[13] = TZR(T) * TdPS(Td)
  ternp[14] = TZR(T) * TdPB(Td)

  ternp[15] = TPS(T) * TdNB(Td)
  ternp[16] = TPS(T) * TdNS(Td)
  ternp[17] = TPS(T) * TdZR(Td)
  ternp[18] = TPS(T) * TdPS(Td)
  ternp[19] = TPS(T) * TdPB(Td)

  ternp[20] = TPB(T) * TdNB(Td)
  ternp[21] = TPB(T) * TdNS(Td)
  ternp[22] = TPB(T) * TdZR(Td)
  ternp[23] = TPB(T) * TdPS(Td)
  ternp[24] = TPB(T) * TdPB(Td)


  center=[-100,-70,-50,-30,-20, -80,-60,-40,-20,-5, -30,-20,0,20,30, 5,20,40,60,80, 20,30,50,70,100]


  numerator = np.sum(center * ternp)
  denonpinator = np.sum(ternp)
  u = numerator/denonpinator

  return u
```

```python
def simulation(t,td):

  def calculate(i):
    N= g * np.sin(x1[i-1]) - ((m*l*x2[i-1]**2*np.cos(x1[i-1])*np.sin(x1[i-1]))/M+m)
    D= l*((4/3)-(m*np.cos(x1[i-1]**2))/M+m)
    fb = N / D - ((np.cos(x1[i-1]))/M+m)/D*u[i-1]
    return fb

  x1 = np.zeros(shape=1000)
  x1[0]=t*np.pi/180

  x2 = np.zeros(shape=1000)
  x2[0]=td

  u = np.zeros(shape=1000)
  u[0]=control(x1[0],x2[0])

  t = np.zeros(shape=1000)
  t[0]=x1[0]*180/np.pi

  g=9.8
  M=1
  m=0.1
  l=0.5

  for i in range(1,1000):
    x1[i] = x1[i-1] - x1[i-1] * 0.01
    x2[i] = calculate(i) * 0.01
    u[i] = control(x1[i-1],x2[i-1])

  for i in range(1,1000):
    t[i] = x1[i]*180/np.pi

  plt.plot(t,'b-')
  plt.show()
  plt.plot(x2,'b-')
  plt.show()
  plt.plot(u,'b-')
  plt.show()

simulation(45,0)
```
